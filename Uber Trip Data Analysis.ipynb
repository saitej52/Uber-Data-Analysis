# Converting the date and time to pandas format

```python
data['Date/Time'] = data['Date/Time'].map(pd.to_datetime)
```

```python
data.head()
```

```python
data.tail()
```

# Feature Engineering

```python
def get_dom(dt):
    return dt.day
```

```python
data['dom'] = data['Date/Time'].map(get_dom)
```

```python
data.head()
```

```python
data.tail()
```

```python
def get_weekday(dt):
    return dt.dayofweek
```

```python
data['Weekday'] = data['Date/Time'].map(get_weekday)
```

```python
data.head()
```

```python
def get_hour(dt):
    return dt.hour
```

```python
data['Hour'] = data['Date/Time'].map(get_hour)
```

```python
data.tail()
```

```python
data.head()
```

# Analyze Date Of Month (dom)

```python
hist(data['dom']) # This gives only ten generalised values of the complete month
```

```python
hist(data['dom'], bins = 30) # But we nedd all the exact values of days in a month. So we would give the bins of 30 for 30 days
```

```python
hist(data['dom'], bins =30, rwidth = 0.9)
```

```python
hist(data['dom'], bins =30, rwidth = .8, range = (0.5,30.5))
xlabel("Date Of Month")
ylabel("Frequency")
title("Frequency of data through Date Of Month")
```
